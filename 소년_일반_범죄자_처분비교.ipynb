# 소년_일반_범죄자_처분비교

```python
import pandas as pd
```

```python
# 전체 범죄자 처분 데이터 로딩
df = pd.read_csv('./범죄자 처분결과.csv', encoding='cp949')
df.head()
```

```python
# 각 분류 별 계 항목 제거
df = df[(df['분류'] != '분류') & (df['분류'] != '계') & (~df['분류.1'].str.contains('계')) & (~df['분류.2'].str.contains('계')) & (~df['분류.3'].str.contains('계'))]

# 기소, 불기소 데이터만 분류에 따라 조합
df_all = pd.DataFrame(data= { '분류': df['분류.1'], '일반 기소': df['계.1'], '일반 불기소': df['계.9'] })

# 기소, 불기소 문자열 -> 숫자로 변경
df_all['일반 기소'] = df_all['일반 기소'].str.replace(',', '').astype(float)
df_all['일반 불기소'] = df_all['일반 불기소'].str.replace(',', '').astype(float)

df_all.head()
```

```python
# 분류를 기준으로 합산
df_all = df_all.groupby(by='분류').sum()

df_all
```

```python
# 기소율 계산
df_all['기소율'] = (df_all['일반 기소'] / (df_all['일반 기소'] + df_all['일반 불기소']) * 100)

df_all
```
